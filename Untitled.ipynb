```python
import os
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
pd.set_option('display.max_colwidth', None)

# Funzione per leggere tutti i file Excel nella cartella "moduli_iscrizione" e unirli in un unico DataFrame
def unisci_excel_da_cartella(cartella):
    dfs = []  # Lista per contenere i DataFrame
    for filename in os.listdir(cartella):
        if filename.endswith('.xlsx'):
            filepath = os.path.join(cartella, filename)
            df = pd.read_excel(filepath)
            dfs.append(df)
    return pd.concat(dfs, ignore_index=True)

# Utilizzo della funzione
cartella = "moduli_iscrizione"
df = unisci_excel_da_cartella(cartella)

```

```python
dizionario_df = {}

for _, gruppo_anno in df.groupby('Anno_Nascita'):
    for _, gruppo_cintura in gruppo_anno.groupby('Cintura'):
        percorso_df = gruppo_cintura[gruppo_cintura['Kata_Percorso'] == 'PERCORSO']
        normale_df = gruppo_cintura[gruppo_cintura['Kata_Percorso'] != 'PERCORSO']

        for sub_df, suffix in zip([normale_df, percorso_df], ["", " PERCORSO"]):
            if len(sub_df) == 0:
                continue

            title = f"{sub_df['Anno_Nascita'].iloc[0][:8]} - {sub_df['Cintura'].iloc[0].replace(', ', '-').replace('/', '')}{suffix}"

            if len(sub_df) > 10:
                pool1, pool2 = [], []
                sub_df = sub_df.sort_values(by='Società')

                for _, atleta in sub_df.iterrows():
                    if len(pool1) <= len(pool2):
                        pool1.append(atleta)
                    else:
                        pool2.append(atleta)

                pool1_df = pd.DataFrame(pool1)
                pool2_df = pd.DataFrame(pool2)

                for pool_num, pool_df in enumerate([pool1_df, pool2_df], 1):
                    pool_title = f"{title} POOL-{pool_num}"
                    pool_df['NOME KATA'] = ''
                    pool_df['1'] = ''
                    pool_df['2'] = ''
                    pool_df['3'] = ''
                    pool_df['4'] = ''
                    pool_df['5'] = ''
                    pool_df['TOT'] = ''
                    dizionario_df[pool_title] = pool_df
            else:
                sub_df['NOME KATA'] = ''
                sub_df['1'] = ''
                sub_df['2'] = ''
                sub_df['3'] = ''
                sub_df['4'] = ''
                sub_df['5'] = ''
                sub_df['TOT'] = ''
                dizionario_df[title] = sub_df
```

```python
import os
import openpyxl
from openpyxl.styles import Alignment, Border, Side, PatternFill, Font
from openpyxl.worksheet.page import PageMargins
from openpyxl.utils import get_column_letter

# Definisci i colori di sfondo
header_fill = PatternFill(start_color="D3D3D3", end_color="D3D3D3", fill_type="solid")
yellow_fill = PatternFill(start_color="FFFFCC", end_color="FFFFCC", fill_type="solid")

# Font per il titolo
title_font = Font(size=14, bold=True)
hair_border = Border(left=Side(style='hair'), right=Side(style='hair'), top=Side(style='hair'), bottom=Side(style='hair'))

# Crea la cartella 'risultato' se non esiste
if not os.path.exists('risultato'):
    os.makedirs('risultato')

for title, x in dizionario_df.items():
    file_name = f"risultato/{title}.xlsx"

    x = x.copy().drop(columns=['Anno_Nascita', 'Cintura', 'Kata_Percorso'])
    x = x.sample(frac=1).sample(frac=1).sample(frac=1).sample(frac=1).sample(frac=1)
    
    # Salva con offset di 2 righe per il titolo e la riga vuota
    x.to_excel(file_name, sheet_name='Foglio1', startrow=2, index=True)

    # Apri il file Excel e seleziona il foglio di lavoro
    wb = openpyxl.load_workbook(file_name)
    ws = wb['Foglio1']

    # Imposta varie opzioni di pagina
    ws.page_setup.orientation = ws.ORIENTATION_LANDSCAPE
    ws.page_margins = PageMargins(left=0.2, right=0.2, top=0.8, bottom=0.2)
    ws.page_setup.fitToPage = True
    ws.page_setup.fitToHeight = 1
    ws.page_setup.fitToWidth = 1

    # Inserisci il titolo e unisci le celle
    ws['A1'] = title
    ws['A1'].font = title_font
    ws['A1'].alignment = Alignment(horizontal='center')
    ws.merge_cells(start_row=1, start_column=1, end_row=1, end_column=ws.max_column)

    # Loop per ogni cella
    for row in ws.iter_rows(min_row=3):  # Inizia dalla 3ª riga
        for cell in row:
            # Imposta l'altezza delle righe a 30
            ws.row_dimensions[cell.row].height = 30

            # Imposta l'allineamento
            if cell.column == 1 or (cell.column >= 6 and cell.column <= 11):
                cell.alignment = Alignment(vertical='center', horizontal='center')
            else:
                cell.alignment = Alignment(vertical='center', horizontal='left')

            if cell.row > 3 and cell.column > 1:
                cell.border = hair_border
        
            # Imposta il colore di sfondo per l'header (3ª riga)
            if cell.row == 3:
                cell.fill = header_fill

            # Imposta il bordo per la colonna J (10ª colonna)
            if cell.column == 11:
                if cell.row == 3:
                    cell.border = Border(top=Side(style='thin'), left=Side(style='thin'), right=Side(style='thin'), bottom=Side(style='thin'))
                elif cell.row == ws.max_row:
                    cell.border = Border(bottom=Side(style='thin'), left=Side(style='thin'), right=Side(style='thin'))
                else:
                    cell.border = Border(left=Side(style='thin'), right=Side(style='thin'), bottom=Side(style='hair'), top=Side(style='hair'))

            # Imposta i colori di sfondo alternati per le righe
            if cell.row > 3 and cell.row % 2 == 1:
                cell.fill = yellow_fill

    # Imposta l'area di stampa e la larghezza delle colonne
    max_row = ws.max_row
    max_col = get_column_letter(ws.max_column)
    ws.page_setup.print_area = f"A1:{max_col}{max_row}"
    ws.page_setup.paperSize = ws.PAPERSIZE_A3  # Imposta la dimensione del foglio in A3
    ws.column_dimensions['A'].width = 8.43
    ws.column_dimensions['B'].width = 45.57
    ws.column_dimensions['C'].width = 24.87
    ws.column_dimensions['D'].width = 24.87
    ws.column_dimensions['E'].width = 45.57

    # Salva le modifiche
    wb.save(file_name)
```
