```python
import pandas as pd
from sqlalchemy import create_engine, Float, VARCHAR, Text
from sqlalchemy.orm import sessionmaker
from sqlalchemy.exc import SQLAlchemyError

pd.set_option("display.max_columns", None)
df = pd.read_csv('iscrizioni.csv',delimiter=';')
df.index.name = 'id_iscrizione'
df.head(2)
```

```python
# Parametri DB
hostname = 'localhost'
username = 'root'
password = '932197'
database_name = 'localdb'

engine = create_engine(f'mysql+pymysql://{username}:{password}@{hostname}/{database_name}')
Session = sessionmaker(bind=engine)
session = Session()
try:
    df.to_sql('iscrizioni', con=engine, if_exists='replace', index=True, method='multi')
except SQLAlchemyError as e:
    print(f"Si è verificato un errore: {e}")
    session.rollback()
finally:
    engine.dispose()
```
